```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 設定參數
np.random.seed(42)
n_series = 500
length = 100
output_dir = "stable_series_images"
os.makedirs(output_dir, exist_ok=True)

# 產生與儲存圖像
for i in range(n_series):
    base = np.random.normal(loc=0.0, scale=0.2, size=length)
    noise = np.random.normal(loc=0.0, scale=0.05, size=length)
    series = np.cumsum(base + noise)

    plt.figure(figsize=(3, 3))
    plt.plot(series, color="black", linewidth=2)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(f"{output_dir}/stable_{i:03d}.png", dpi=100, bbox_inches='tight', pad_inches=0)
    plt.close()
import subprocess
subprocess.run('zip -r stable_series_images.zip stable_series_images', shell=True, check=True)
```
